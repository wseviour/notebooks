Does the SOI/SAM relationship show decadal variability?
=======================================================

```python
% matplotlib inline 
import pandas as pd
import numpy as np
import scipy.signal as signal
```

```python
soi = pd.read_excel('./data/SOI_index.xls', 'Sheet1')
```

```python
soi
```

```python
sam = pd.read_excel('./data/BAS_SAM.xls','Sheet1')
```

```python
sam
```

```python
sam_oct_nov_all = (sam['OCT']+sam['NOV'])/2.
soi_jun_jul_all = (soi['JUN']+soi['JUL'])/2.
```

```python
sam_oct_nov = sam_oct_nov_all.ix[1957:2012]
soi_jun_jul = soi_jun_jul_all.ix[1957:2012]
```

```python
np.corrcoef(sam_oct_nov.values, soi_jun_jul.values)
```

```python
sam_oct_nov_det = signal.detrend(sam_oct_nov.values)
soi_jun_jul_det = signal.detrend(soi_jun_jul.values)
```
